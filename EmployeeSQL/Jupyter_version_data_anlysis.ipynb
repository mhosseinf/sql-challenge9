```python
import pandas as pd
```

```python
# path to CSV file.
csv_file_path1 = 'resource/departments.csv'
# Read the CSV file into a DataFrame
departments_df = pd.read_csv(csv_file_path1)
# Display the DataFrame
departments_df.head()
```

```python
# path to CSV file.
csv_file_path2 = 'resource/dept_emp.csv'
# Read the CSV file into a DataFrame
dept_emp_df = pd.read_csv(csv_file_path2)
# Display the DataFrame
dept_emp_df.head()
```

```python
# path to CSV file.
csv_file_path3 = 'resource/dept_manager.csv'
# Read the CSV file into a DataFrame
dept_manager_df = pd.read_csv(csv_file_path3)
# Display the DataFrame
dept_manager_df.head()
```

```python
# path to CSV file.
csv_file_path4 = 'resource/employees.csv'
# Read the CSV file into a DataFrame
employees_df = pd.read_csv(csv_file_path4)
# Display the DataFrame
employees_df.head()
```

```python
# path to CSV file.
csv_file_path5 = 'resource/salaries.csv'
# Read the CSV file into a DataFrame
salaries_df = pd.read_csv(csv_file_path5)
# Display the DataFrame
salaries_df.head()
```

```python
# path to CSV file.
csv_file_path6 = 'resource/titles.csv'
# Read the CSV file into a DataFrame
titles_df = pd.read_csv(csv_file_path6)
# Display the DataFrame
titles_df.head()
```

1-List the employee number, last name, first name, sex, and salary of each employee

```python
# Merge employees_df and salaries_df to get employee details with salaries
employee_salaries_df = employees_df.merge(salaries_df, on='emp_no')

# Select the desired columns
employee_info_df = employee_salaries_df[['emp_no', 'last_name', 'first_name', 'sex', 'salary']]

# Display the result
employee_info_df
```

2-List the first name, last name, and hire date for the employees who were hired in 1986.

```python
# Convert the 'hire_date' column to a datetime format
employees_df['hire_date'] = pd.to_datetime(employees_df['hire_date'])

# Filter employees hired in 1986
employees_hired_1986_df = employees_df[employees_df['hire_date'].dt.year == 1986]

# Select the desired columns
result_df = employees_hired_1986_df[['first_name', 'last_name', 'hire_date']]

# Display the result
result_df
```

3-List the manager of each department along with their department number, department name, employee number, last name, and first name.

```python
# Merge the departments_df and dept_manager_df to get department names and managers
department_manager_df = pd.merge(departments_df, dept_manager_df, on='dept_no')

# Merge the department_manager_df with employees_df to get manager details
manager_details_df = pd.merge(department_manager_df, employees_df, left_on='emp_no', right_on='emp_no')

# Select the desired columns
result_df = manager_details_df[['dept_no', 'dept_name', 'emp_no', 'last_name', 'first_name']]

# Rename the columns for clarity
result_df.columns = ['department_number', 'department_name', 'employee_number', 'last_name', 'first_name']

# Display the result
result_df
```

4-List the department number for each employee along with that employee’s employee number, last name, first name, and department name.

```python
# Merge the employees_df and dept_emp_df to get employee details and department numbers
employee_department_df = pd.merge(employees_df, dept_emp_df, on='emp_no')

# Merge the employee_department_df with departments_df to get department names
employee_department_df = pd.merge(employee_department_df, departments_df, left_on='dept_no', right_on='dept_no')

# Select the desired columns
result_df = employee_department_df[['emp_no', 'last_name', 'first_name', 'dept_no', 'dept_name']]

# Rename the columns for clarity
result_df.columns = ['employee_number', 'last_name', 'first_name', 'department_number', 'department_name']

# Display the result
result_df
```

5-List the first name, last name, and sex of each employee whose first name is Hercules and whose last name begins with the letter B

```python
# Filter employees whose first name is Hercules and last name begins with 'B'
filtered_employees_df = employees_df[(employees_df['first_name'] == 'Hercules') & (employees_df['last_name'].str.startswith('B'))]

# Select the desired columns
result_df = filtered_employees_df[['first_name', 'last_name', 'sex']]

# Display the result
result_df
```

6-List each employee in the Sales department, including their employee number, last name, and first name.

```python
# Merge the employees_df and dept_emp_df dataframes to get department information
merged_df = pd.merge(employees_df, dept_emp_df, on='emp_no')

# Merge the merged_df with departments_df to get department names
sales_employees_df = pd.merge(merged_df, departments_df, left_on='dept_no', right_on='dept_no')

# Filter employees in the Sales department
sales_employees_df = sales_employees_df[sales_employees_df['dept_name'] == 'Sales']

# Select the desired columns
result_df = sales_employees_df[['emp_no', 'last_name', 'first_name']]

# Display the result
result_df
```

7-List each employee in the Sales and Development departments, including their employee number, last name, first name, and department name.

```python
# Merge the employees_df and dept_emp_df dataframes to get department information
merged_df = pd.merge(employees_df, dept_emp_df, on='emp_no')

# Merge the merged_df with departments_df to get department names
all_employees_df = pd.merge(merged_df, departments_df, left_on='dept_no', right_on='dept_no')

# Filter employees in the Sales and Development departments
sales_dev_employees_df = all_employees_df[
    (all_employees_df['dept_name'] == 'Sales') | (all_employees_df['dept_name'] == 'Development')
]

# Select the desired columns
result_df = sales_dev_employees_df[['emp_no', 'last_name', 'first_name', 'dept_name']]

# Display the result
result_df
```

8-List the frequency counts, in descending order, of all the employee last names (that is, how many employees share each last name).

```python
# Count the frequency of each last name in the employees_df dataframe
last_name_counts = employees_df['last_name'].value_counts().reset_index()

# Rename the columns for clarity
last_name_counts.columns = ['last_name', 'frequency']

# Sort the dataframe by frequency in descending order
last_name_counts = last_name_counts.sort_values(by='frequency', ascending=False)

# Display the result
last_name_counts
```
